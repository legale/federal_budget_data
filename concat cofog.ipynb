```python
cofog = pd.read_excel('cofog_rus_mod2.xlsx')

c1 = cofog[['code','name','name']].dropna()
c2 = cofog[['code','name','name2']].dropna()
c3 = cofog[['code','name','name3']].dropna()
c4 = cofog[['code','name','name4']].dropna()
c5 = cofog[['code','name','name5']].dropna()
```

```python

cols = ['code','name','name2']
c1.columns = cols
c2.columns = cols
c3.columns = cols
c4.columns = cols
c5.columns = cols
cofog = pd.concat([c1,c2,c3,c4,c5],axis=0,ignore_index=True)
```
