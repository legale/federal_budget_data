```python
data = pp.load('02-19.pkl')
cofog_ = pd.read_excel('cofog_rus_mod3.xlsx', dtype = {'code': str, 'name': str, \
                                                      'name2': str, 'code2': str, \
                                                      'div':str, 'subdiv':str})
cofog = pd.read_excel('cofog_rus_mod1.xlsx')
cofog['code2'] = cofog['code'].astype(str).str.ljust(4, '0')
cofog[['div','subdiv']] = cofog['code2'].str.extract(r'\d(\d\d)(\d)')
cofog = cofog.astype({'div': str, 'subdiv': str})
```

```python
writer = pd.ExcelWriter('02-19.xlsx')
yy = {}
for y in data:
    df = data[y]
    df = pp.levenstein_merge(df,cofog_,'Наименование','name2',60)
    yy[y] = df[['div','РЗ','code','Сумма']]
    yy[y].to_excel(writer,y, index=False)

cofog_.to_excel(writer,'cofog', index=False)
writer.save()
    
```

```python
def manuscript(yy):
    m = pd.DataFrame([],columns=['div'])
    for y in yy:
        tomerge = yy[y][['div', 'Сумма']].groupby(['div']).agg({'Сумма': sum})
        tomerge.columns = [y]
        m = m.merge(tomerge, on='div', how='outer')
    return m

m = manuscript(yy)
```
