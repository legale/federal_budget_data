```python
#yy = pp.load('02-19.pkl')
yy = pd.read_excel('budget_data_raw_02-19.xlsx', dtype=str)
yy.head(2)
```

```python
def prep_data(df):
    lst = []
    for i in df.index:
        row = df.iloc[i,:]
        div = row[0]
        name = row[1]
        
        #skip first 2 columns
        row_ = row[2:]
        for k, y in enumerate(row_.index):
            begin = float(row_[k]) # current year
            try:
                end = float(row_[k + 1]) # next year
            except:
                end = float(row_[k])
            
            
            range_ = end - begin
            step = range_ / 10

            cur = begin
            for n in range(0,10):
                last = cur
                cur = begin + step * (n+1)
                lst.append({'name': div + ' ' + name,
                            'value': cur,
                            'year': float(str(y) + '.' + str(n)),
                            'lastValue': last,
                            'rank': 0})
    return lst
```

```python
data = prep_data(yy)
```

```python
dfp = pd.DataFrame(data)
dfp.shape
```

```python
dfp_ = dfp.sort_values(by=['year','value'])
dfp_.reset_index(drop=True, inplace = True)
dfp_['rank'] = range(1,1801)

dfp_
```
